```python

x = np.arange(0, 10, 0.1)
y = np.sin(x)

plt.plot(x, y)

plt.title("Sins Grafii")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(75, 250, 50)

plt.scatter(x, y, c=colors, s=sizes)

plt.title("Point Chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python

data = np.random.randn(500)

plt.hist(data, bins=50)

plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(5, 10, 5)

plt.bar(x, y)

plt.title("Column Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python

x = np.arange(-10, 10, 0.10)
y = np.arange(-10, 10, 0.10)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**2 + y**2)
z = np.sin(r)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z)

ax.set_title("3D Graphics")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```
